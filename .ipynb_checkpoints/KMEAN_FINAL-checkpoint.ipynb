```python
#importer les packages nécessaires
import pandas as pd
import numpy as np

#les packages de visualisation
import seaborn as sns
import matplotlib.pyplot as plt

#importer les packages de préparation de données
from sklearn.preprocessing import StandardScaler

#importer les packages de clustering
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
#importer les données
df = pd.read_excel(r'C:\Users\ECC\Desktop\PFE\Toddler_Autism_dataset_CLEAN.xlsx')

df
```

```python
#centrer et reduire nos données
scaler = StandardScaler()
x_scaled = scaler.fit_transform(df)

x_scaled
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)

wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.title('la méthode du coude pour determiner le nombre des clusters')
plt.xlabel('nombre de clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Perform PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)

# Perform K-means clustering
kmeans_optimal = KMeans(n_clusters=3, init='k-means++', random_state=42)
clusters_optimal = kmeans_optimal.fit_predict(x_scaled)

# Ajoutez le résultat des clusters aux données pour analyse ultérieure
df['cluster'] = clusters_optimal

# Visualiser les clusters
plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=clusters_optimal, cmap='viridis')
plt.title('Visualisation des clusters K-means (k=3)')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Add the cluster result to the original data for future analysis
df['cluster'] = clusters_optimal
df['cluster'] 
```

```python
# Analyze Cluster Centroids
centroids = kmeans_optimal.cluster_centers_
centroids_pca = pca.transform(centroids)

# Plot Centroids in PCA Space
plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=clusters_optimal, cmap='viridis', alpha=0.5)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='red', marker='X', s=200)
plt.title('Cluster Centroids in PCA Space')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Principal Components Analysis
components = pca.components_
explained_variance = pca.explained_variance_ratio_

# Display PCA Component Loadings
components_df = pd.DataFrame(components, columns=df.columns[:-1])  # Exclude 'cluster' column
print("PCA Component Loadings:")
print(components_df)
```

```python
# Display Explained Variance Ratio
explained_variance_df = pd.DataFrame(explained_variance, index=['Component 1', 'Component 2'], columns=['Explained Variance'])
print("Explained Variance Ratio:")
print(explained_variance_df)
```

```python
# Cluster Profiling
cluster_profile = df.groupby('cluster').mean()
print("Cluster Profiling:")
print(cluster_profile)
```

```python
# Visualizing Cluster Profiles
for column in df.columns[:-1]:  # Exclude 'cluster' column
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='cluster', y=column, data=df)
    plt.title(f'{column} by Cluster')
    plt.show()
```
